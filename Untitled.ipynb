```python
import matplotlib.pyplot as plt

# Room dimensions
room_length, room_width = 10, 8  # meters

# Device placement
speakers = [(2, 4), (8, 4)]
display = [(5, 6)]

# Visualization
plt.figure(figsize=(8, 6))
plt.plot([0, room_length, room_length, 0, 0], [0, 0, room_width, room_width, 0], 'b-')
for x, y in speakers:
    plt.plot(x, y, 'ro', label='Speaker')
plt.plot(*display[0], 'go', label='Display')
plt.xlabel('Length (m)')
plt.ylabel('Width (m)')
plt.title('Room Layout with AV Device Placement')
plt.legend()
plt.show()
```
